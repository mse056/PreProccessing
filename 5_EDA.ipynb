```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data/Automobile.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
sns.distplot(data['wheel_base'])
plt.show()
```

```python
sns.distplot(data['price'])
plt.show()
```

```python
sns.jointplot(data=data, x='price', y='horsepower')
plt.show()
```

```python
sns.jointplot(data=data, x='price', y='horsepower', kind='hex')
plt.show()
```

```python
sns.pairplot(data)
plt.show()
```

```python
sns.swarmplot(data=data, x='fuel_type', y='horsepower')
plt.show()
```

```python
sns.boxplot(data=data, x='number_of_doors', y='horsepower', hue=data['fuel_type'])
plt.show()
```

```python
sns.barplot(data=data, x='number_of_doors', y='horsepower', hue=data['fuel_type'])
plt.show()
```

```python
sns.barplot(data=data, x='number_of_doors', y='horsepower', hue=data['engine_location'])
plt.show()
```
