```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data/games.csv')
data.head()
```

```python
data.info()
```

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True) # all rows which have at at least one column null will be droped
```

```python
data.isnull().sum()
```

#### if you think data hasn't any more null you 're wrong, let see...

```python
sns.histplot(data['average_rating'])
plt.show()
```

#### as you can see this isn't a normal data and we have many zero values

```python
sns.distplot(data['yearpublished'])
plt.show()
```

```python
sns.jointplot(data= data, x='minage', y='average_rating')
plt.show()
```

```python
sns.jointplot(data= data, x='minage', y='total_owners')
plt.show()
```

```python
data2 = data[data['minage'] < 20]
sns.jointplot(data= data2, x='minage', y='total_owners')
plt.show()
```

```python
sns.stripplot(data=data, x='type', y='playingtime')
plt.show()
```

```python
sns.regplot(data=data, x='playingtime', y='average_rating')
plt.show()
```
